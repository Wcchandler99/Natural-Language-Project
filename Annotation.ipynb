```python
import os
from openai import OpenAI


#Load your text file
with open('TinyStoriesV3-GPT4-train.txt', 'r', encoding='utf-8') as f:
    text_data = f.read()


def segment_text(text):
    raw_paragraphs = text.split('\n')
    stories = [p.strip() for p in raw_paragraphs if p.strip()]
    return stories


stories = segment_text(text_data)

print(stories[0])
```

```python
system_prompt = "Return the given story segmented into beginning, rising action, climax, and falling action in JSON format. Do not edit the story at all return the exact text."

annotated_stories = []

#os.environ["OPENAI_API_KEY"] 

i = 1

for story in stories:
    print(i)
    prompt = story

    client = OpenAI(
        # This is the default and can be omitted
        api_key=os.environ.get("OPENAI_API_KEY"),
    )

    response = client.responses.create(
        model="gpt-4o",
        instructions=system_prompt,
        input=prompt,
    )

    annotated_stories.append(response.output_text)

    i += 1
```

```python
print(annotated_stories)
```
